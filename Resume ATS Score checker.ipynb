```python
import os
# Define the directory containing your CSV files
folder_path = r"C:\Users\Prakriti\.cache\kagglehub\datasets\suriyaganesh\resume-dataset-structured\versions\2"

# List all files in the directory
files = os.listdir(folder_path)

# Filter only the CSV files
csv_files = [file for file in files if file.endswith('.csv')]

# Load each CSV file into a pandas DataFrame
data_frames = {file: pd.read_csv(os.path.join(folder_path, file)) for file in csv_files}

# Display the first few rows of each DataFrame
for file, df in data_frames.items():
    print(f"Data from {file}:")
    print(df.sample(50))
    print("\n")
```
